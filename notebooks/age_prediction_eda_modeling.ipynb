# NHANES Age Prediction: Exploratory Data Analysis & Modeling
This notebook contains the complete workflow for processing the NHANES health data, performing stratified evaluation, and building tree-based classifiers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Load datasets
train = pd.read_csv('../data/Train_dataset (1).csv')
test = pd.read_csv('../data/Test_dataset (1).csv')

print(f'Train Shape: {train.shape} | Test Shape: {test.shape}')
```

## 1. Target Value Analysis

```python
# Encode categorical strings to numeric target metrics
train['target'] = train['age_group'].apply(lambda x: 1 if x == 'Senior' else 0)

sns.countplot(x='age_group', data=train, palette='viridis')
plt.title('Target Class Distribution (Adult vs Senior)')
plt.show()
```

## 2. Feature Preprocessing Pipeline

```python
X_train = train.drop(columns=['SEQN', 'age_group', 'target'])
y_train = train['target']
X_test = test.drop(columns=['SEQN'])

num_cols = ['BMXBMI', 'LBXGLU', 'LBXGLT', 'LBXIN']
cat_cols = ['RIAGENDR', 'PAQ605', 'DIQ010']

# Apply strategic imputations to protect validation consistency
num_imputer = SimpleImputer(strategy='median')
X_train[num_cols] = num_imputer.fit_transform(X_train[num_cols])
X_test[num_cols] = num_imputer.transform(X_test[num_cols])

cat_imputer = SimpleImputer(strategy='most_frequent')
X_train[cat_cols] = cat_imputer.fit_transform(X_train[cat_cols])
X_test[cat_cols] = cat_imputer.transform(X_test[cat_cols])
```

## 3. Local Model Evaluation Framework

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

baseline_rf = RandomForestClassifier(random_state=42)
rf_cv = cross_val_score(baseline_rf, X_train, y_train, cv=skf, scoring='roc_auc')
print(f'Random Forest Baseline Mean Cross-Validation ROC-AUC: {rf_cv.mean():.4f}')

xgb_model = XGBClassifier(random_state=42, n_estimators=100, learning_rate=0.05)
xgb_cv = cross_val_score(xgb_model, X_train, y_train, cv=skf, scoring='roc_auc')
print(f'XGBoost Optimized Model Mean Cross-Validation ROC-AUC: {xgb_cv.mean():.4f}')
```
